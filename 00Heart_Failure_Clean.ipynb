```python
def age_Range(val):
    if val > 25 and val <= 35:
        return '25-35'
    elif val > 35 and val <= 45:
        return '36-45'
    elif val > 45 and val <= 55:
        return '46-55'
    elif val > 55 and val <= 65:
        return '56-65'
    else:
        return '66-80'
```

```python
newHeart = heart.copy()
# newHeart['Age_Ranges'] = newHeart.loc[(heart.Age.values > 25) & (heart.Age.values <= 35), 'Age_Ranges'] = '25-35'
# newHeart['Age_Ranges'] = newHeart.loc[(heart.Age.values > 35) & (heart.Age.values <= 45), 'Age_Ranges'] = '36-45'
# newHeart['Age_Ranges'] = newHeart.loc[(heart.Age.values > 45) & (heart.Age.values <= 55), 'Age_Ranges'] = '46-55'
# newHeart['Age_Ranges'] = newHeart.loc[(heart.Age.values > 55) & (heart.Age.values <= 65), 'Age_Ranges'] = '56-65'
# newHeart['Age_Ranges'] = newHeart.loc[(heart.Age.values > 65) & (heart.Age.values <= 80), 'Age_Ranges'] = '66-80'
newHeart['Age_Ranges'] = newHeart['Age'].apply(age_Range)
newHeart.head(10)
```

```python
newHeart['Gender'] = newHeart['Sex'].apply(lambda x: 0 if x == 'M' else 1)
newHeart.head()
```

```python
newHeart['ExerciseAngina_num'] = newHeart['ExerciseAngina'].apply(lambda x: 0 if x == 'N' else 1)
```
